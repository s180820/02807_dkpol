```python
wnl = nltk.WordNetLemmatizer() 

tokens_list = []
for i in range(data.shape[0]):
    raw = data["text"].iloc[i]
    tokensraw = word_tokenize(raw) #get tokens 
    tokens = [word.lower() for word in tokensraw if word.isalnum()] #only get words and numbers
    tokens = [wnl.lemmatize(t) for t in tokens] #lemmatize tokens
    tokens_list += tokens
```

```python
#Function for removing stop words and make words lowercase
def rem_stop(textlist):
    stopwords = nltk.corpus.stopwords.words('danish')
    realdeal = [w for w in textlist if w.lower() not in stopwords]
    return realdeal
```

```python
course_tokens = rem_stop(tokens_list) 

#In addition also remove specified words
rem_words = ['rt', 'http'] 
course_tokens = list(filter(lambda a: a not in rem_words, course_tokens))

pop_list = nltk.FreqDist(ch.lower() for ch in course_tokens if ch.isalpha())
pop_list
```
